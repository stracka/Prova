```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

gains = np.array([[62 , 245.3485551204462  , 29803.6469494489  , 29286.468300484838],
                  [65 , 326.95480690425927 , 39232.01238609259 , 38554.28688393648],
                  [69 , 433.20030364762556  , 52348.28738311609 , 51395.694403342466],
                  [73 , 541.5964775086505  , 66949.8893783407 , 65682.28302913216]])

y = gains[:,0]
x = gains[:,3]

res = stats.linregress(x, y)


plt.plot(x, y, 'o', label='original data')

plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')

plt.legend()

plt.show()

Vbd =  res.intercept 
print('Vbd =', res.intercept, '+/-', res.intercept_stderr)
```

```python
# APA and LU
ict = np.array([[62 , 0.14023718097600057 ,  0.12298948264076326],
                  [65 , 0.27408473763956764 ,  0.21512284822385563],
                  [69 , 0.486100795451476 , 0.327098132871801 ],
                  [73 , 0.6833079525970864 , 0.4059316368955822 ]])

x = (ict[:,0]-Vbd)/2
apa = ict[:,1]
lam = ict[:,2]

plt.grid(True,'both','both')

plt.plot(x, apa, 'o', label='APA')

plt.plot(x, lam, 'o', label='lambda')

plt.legend()

plt.show()
```

```python
# P(AP in [t,8000]ns) with t = 80ns or 160ns

ap = np.array([[62 , 0.035, 0.071],
                [65 , 0.040, 0.091],
                [69 , 0.041, 0.098],
                [73 , 0.045, 0.102]])


x = (ap[:,0]-Vbd)/2
ap160 = ap[:,1]
ap80 = ap[:,2]


plt.grid(True,'both','both')

plt.plot(x, ap160, 'o', label='t>160')
plt.plot(x, ap80, 'o', label='t>80')


plt.legend()

plt.show()
```
